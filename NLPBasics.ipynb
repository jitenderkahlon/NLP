```python
import nltk
nltk.download('punkt_tab')
```

```python
corpus = '''The food is good and the boy's is a good boy'''
vocabulary = ['the', 'food', 'is', 'good', 'and', 'boy', 'a']
```

```python
# Tokenization 

from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
```

```python
tokens = word_tokenize(corpus)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# Text cleaning 
for i in range(len(tokens)):
    tokens[i] = tokens[i].lower()
```

```python
list(set(tokens))
```

```python
for token in list(set(tokens)):
    if token in stopwords.words('english'):
        tokens.remove(token)
```

```python
words = ['eat', 'eating', 'play', 'playing', 'playful','player', 'finally', 'history', 'congratulations']
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
st = PorterStemmer()
ss = SnowballStemmer('english')
```

```python
# Stemming 
print ("-------Porter Stemmer -------")
for word in words:
    print (f"{word} --------------- {st.stem(word)}")

print ("\n-------Snowball Stemmer -------\n")
for word in words:
    print (f"{word} --------------- {ss.stem(word)}")
```

```python
lt = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
words = ['eat', 'eating', 'play', 'playing', 'playful','player', 'finally', 'history', 'congratulations', 'historical']
for word in words:
    print (f"{word} --------------- {lt.lemmatize(word, pos='v')}")
```
